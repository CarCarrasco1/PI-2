### Índice ROI

```python

# convertir la columna de fecha en un objeto de fecha
df['Date'] = pd.to_datetime(df['Date'])

# ordenar el dataframe por fecha ascendente
df.sort_values(by='Date', ascending=False, inplace=True)

# calcular el ROI diario
df['Daily Return'] = df.groupby('Symbol')['Close'].pct_change()

# calcular el ROI acumulado por día
df['ROI'] = (df['Daily Return'] + 1).groupby(df['Symbol']).cumprod()

#Elimino los valores del ultimo dia ya que no hay informacion aun de ellos apra devolver el índice

df.dropna(subset=['ROI'], inplace=True)
```

```python
df.head()
```

El ROI diario puede proporcionar información útil sobre el rendimiento de una inversión a corto plazo. Para sacar conclusiones de un ROI diario, es importante tener en cuenta el contexto de la inversión, como el sector en el que se encuentra la empresa, las condiciones del mercado y los factores económicos y políticos que puedan afectar el rendimiento de la inversión.

Aquí hay algunos puntos a considerar al analizar el ROI diario:

La variabilidad del ROI: El ROI diario puede variar significativamente de un día a otro, por lo que es importante tener en cuenta la variabilidad a lo largo del tiempo y considerar el promedio del ROI diario para obtener una imagen más precisa del rendimiento a corto plazo.

La tendencia general: Es importante examinar la tendencia general del ROI diario para ver si hay una tendencia al alza o a la baja. Esto puede proporcionar información sobre la salud financiera de la empresa y el desempeño de su sector.

La comparación con otros indicadores: Es útil comparar el ROI diario con otros indicadores, como el volumen de operaciones y los cambios en el precio de las acciones, para obtener una imagen más completa del rendimiento de la inversión a corto plazo.

La relación con la estrategia de inversión: El ROI diario debe evaluarse en relación con la estrategia de inversión general. Si la inversión se realiza a largo plazo, el ROI diario puede tener menos importancia que otros indicadores a largo plazo, como el crecimiento de los ingresos y los dividendos.

Es importante recordar que el ROI diario solo proporciona información sobre el rendimiento a corto plazo y no debe utilizarse como la única medida de éxito de una inversión.


### Volatilidad

La volatilidad es una medida importante para evaluar el riesgo y la fluctuación de los precios de los activos financieros en el mercado. 

```python
df.info()
```

```python
# Calculo el retorno diario
df['Daily Return'] = df['Adj_Close'].pct_change()

# Calculo la volatilidad diaria
volatility = df['Daily Return'].std()

# Agrego la columna de volatilidad al df
df['Volatility'] = volatility
```

```python
df.info() #Compruebo la agregación de la columna
```

La volatilidad diaria es una medida de la variabilidad de los precios de un activo en un solo día de negociación. Es una medida importante de riesgo y se utiliza a menudo para determinar el tamaño de las posiciones que un inversor puede tomar en un activo determinado.

Por ejemplo, si la volatilidad diaria de un activo es alta, significa que sus precios cambian rápidamente en un solo día. Esto puede ser una señal de que el activo es muy riesgoso y que un inversor debe ser cauteloso al tomar una posición en él. Por otro lado, si la volatilidad diaria de un activo es baja, significa que sus precios cambian lentamente en un solo día. Esto puede ser una señal de que el activo es menos riesgoso y que un inversor puede tomar una posición más grande en él.

En resumen, la volatilidad diaria es una medida importante de riesgo y puede ayudar a los inversores a tomar decisiones informadas sobre el tamaño de las posiciones que deben tomar en un activo determinado.

Entonces, supongamos que una acción A tiene un precio inicial de $100. Si la volatilidad diaria de la acción A es del 2%, entonces se esperaría que la acción fluctúe en un rango de +/- $2 al día.

En otras palabras, si la volatilidad diaria de la acción A es alta, por ejemplo del 10%, entonces se esperaría que la acción fluctúe en un rango de +/- $10 al día. Esto significa que el precio de la acción puede subir o bajar significativamente en un solo día.

**Por lo tanto, una volatilidad diaria considerada estable dependerá del precio inicial de la acción y del sector en el que se encuentra la empresa. En general, se considera que una volatilidad diaria del 1-2% es relativamente estable para una acción. Sin embargo, las empresas de sectores más volátiles, como las tecnológicas, pueden tener volatilidades diarias más altas y aún considerarse estables.**

A pesar de la volatilidad que presentan las acciones de empresas tecnológicas, muchas personas todavía consideran que invertir en ellas es una buena opción debido a su potencial de crecimiento y a su papel en la economía digital actual.

Las empresas tecnológicas pueden tener una mayor capacidad para innovar, desarrollar nuevos productos y servicios, y adaptarse rápidamente a los cambios del mercado. Además, la pandemia ha acelerado la adopción de tecnologías digitales en muchos aspectos de la vida, lo que ha aumentado la demanda de productos y servicios tecnológicos.

Por supuesto, como en cualquier inversión, hay riesgos asociados. Es importante realizar una investigación exhaustiva sobre las empresas tecnológicas antes de invertir en ellas y considerar aspectos como su modelo de negocio, su posición en el mercado, su equipo de liderazgo y su historial financiero. También es importante tener en cuenta la diversificación de la cartera de inversiones y el horizonte temporal de la inversión.

### Volumen promedio anual.

```python
df_grouped = df.groupby([df['Symbol'], df['Year']])
df_volume = df_grouped['Volume'].mean()
df_volume = df_volume.to_frame()
df_volume = df_volume.reset_index()
df_volume = df_volume.rename(columns={'Volume': 'Avg_Volume'})
df_volume['Company'] = df_volume['Symbol']
df_volume_anual = df_volume[['Company', 'Year', 'Avg_Volume']]
```

```python
df_volume_anual.sort_values(by='Year', ascending=False, inplace=True)
df_volume_anual.head()
```

El volumen promedio anual indica el promedio diario de acciones que se negociaron durante un año determinado. Este KPI es importante para los inversores ya que puede ayudarles a medir la liquidez de una acción y la facilidad para comprar o vender grandes cantidades de acciones sin afectar significativamente su precio.

Un volumen promedio anual alto puede ser indicativo de una acción popular y en demanda, mientras que un volumen promedio anual bajo puede indicar una acción menos popular o una empresa con una capitalización de mercado menor. Además, los cambios en el volumen promedio anual pueden ser utilizados como una señal para identificar cambios en el interés de los inversores o cambios en la dinámica del mercado.

Un volumen promedio alto puede indicar que hay una gran cantidad de compradores y vendedores activos interesados en la acción en cuestión, lo que generalmente se considera una señal de liquidez y puede indicar que la acción es popular entre los inversores. Un volumen alto también puede hacer que sea más fácil comprar o vender las acciones.
